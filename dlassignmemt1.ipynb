```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# Load MNIST
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

x_train = x_train / 255.0
x_test = x_test / 255.0

# Model
model = models.Sequential([
    layers.Flatten(input_shape=(28,28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(x_train, y_train, epochs=10, validation_split=0.1, batch_size=32)

test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test Accuracy:", test_acc)

# Accuracy Plot
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('MNIST Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train','Validation'])
plt.show()

# Loss Plot
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('MNIST Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train','Validation'])
plt.show()

# Predictions
y_pred = np.argmax(model.predict(x_test), axis=1)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("MNIST Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Classification Report
print("MNIST Classification Report")
print(classification_report(y_test, y_pred))
```

```python
# Load Fashion MNIST
(x_train_f, y_train_f), (x_test_f, y_test_f) = tf.keras.datasets.fashion_mnist.load_data()

x_train_f = x_train_f / 255.0
x_test_f = x_test_f / 255.0

# Model
model_f = models.Sequential([
    layers.Flatten(input_shape=(28,28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model_f.compile(optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])

history_f = model_f.fit(x_train_f, y_train_f, epochs=10, validation_split=0.1, batch_size=32)

test_loss_f, test_acc_f = model_f.evaluate(x_test_f, y_test_f)
print("Test Accuracy:", test_acc_f)

# Accuracy Plot
plt.plot(history_f.history['accuracy'])
plt.plot(history_f.history['val_accuracy'])
plt.title('Fashion MNIST Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train','Validation'])
plt.show()

# Loss Plot
plt.plot(history_f.history['loss'])
plt.plot(history_f.history['val_loss'])
plt.title('Fashion MNIST Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train','Validation'])
plt.show()

# Predictions
y_pred_f = np.argmax(model_f.predict(x_test_f), axis=1)

# Confusion Matrix
cm_f = confusion_matrix(y_test_f, y_pred_f)
plt.figure(figsize=(8,6))
sns.heatmap(cm_f, annot=True, fmt='d', cmap='Blues')
plt.title("Fashion MNIST Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Classification Report
print("Fashion MNIST Classification Report")
print(classification_report(y_test_f, y_pred_f))
```
